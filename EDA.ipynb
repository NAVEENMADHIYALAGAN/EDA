- By using the below digram you can find the average age to fullfill the Null values in the Age columns by define the function
- P1Avg age is37 P2-27, P3-24

```python
plt.figure(figsize=(12,7))

```

```python
sns.boxplot(data=df,x='Pclass',y='Age',palette='winter',hue='Pclass')
```

- Below function check the null values in the Age column and put the avgerage age based on the Pclass 

```python
def impute_age(cols):
    Age = cols[0]  # The 'Age' column
    Pclass = cols[1]  # The 'Pclass' column
    
    if pd.isnull(Age):  # Check if 'Age' is null
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        elif Pclass == 3:
            return 24
    else:
        return Age  # Return the existing age if it's not null
```

```python
df['Age'] = df[['Age', 'Pclass']].apply(impute_age, axis=1)
```

```python
sns.heatmap(df.isnull())
```
